```python
import pandas as pd
data = {
    'horas': [0, 10, 20, 30, 40, 50],
    'vibracion': [0.5, 0.8, 1.5, 3.2, 6.0, 10.2]
}

df = pd.DataFrame(data)
df
```

```python
p_ajustado = Polynomial.fit(df['horas'], df['vibracion'], 2)
print(f'Ecuacion ajustada: {p_ajustado}')
```

```python
prediccion = p_ajustado(60)
print(f'vibracion estimada a las 60h: {prediccion:.2f}') 
```

```python
p_ajustado.deriv()
```
