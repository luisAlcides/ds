### Puntaje Z: estandarizando datos

```python
import numpy as np

datos = np.array([4.5, 2.8, 7.1, 5, 6.3])

media = np.mean(datos)
desviacion = np.std(datos, ddof=1)

zscores = (datos - media) / desviacion
zscores
```

Área entre 4 y 6 horas = porcentaje de personas en ese rango

```python
media = 5
desviacion = 1.5
dato1 = 4
dato2 = 6

z1 = (dato1-media) / desviacion
z3 = (dato2-media) / desviacion
```

```python
z1, z3
```

```python
from scipy.stats import norm

porcentaje = (norm.cdf(z3) - norm.cdf(z1)) * 100
print(f"{porcentaje:.2f}%")
```

```python
import numpy as np
tiempo_celular = np.array([
    4.5, 2.8, 7.1, 5.0, 6.3, 3.2, 5.8, 4.1, 6.5, 3.9,
    7.8, 4.9, 5.3, 2.5, 6.0, 4.8, 3.0, 8.5, 6.7, 5.5,
    4.6, 2.9, 7.3, 5.1, 6.4, 3.3, 5.9, 4.2, 6.6, 4.0,
    7.9, 5.0, 5.4, 2.6, 6.1, 4.7, 3.1, 8.6, 6.8, 5.6,
    5.0, 2.7, 7.5, 5.2, 6.5, 3.4, 6.0, 4.3, 6.7, 4.1
])
```

```python
media = np.mean(tiempo_celular)
desviacion = np.std(tiempo_celular, ddof=1)
n = len(tiempo_celular)

z = 1.96
error_estandar = desviacion / np.sqrt(n)
margen_error = z * error_estandar

limite_inferior = media - margen_error
limite_superior = media + margen_error

print(media, limite_inferior, limite_superior)
```
